# Import The Packages

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
from sklearn import neighbors 
import matplotlib.pyplot as plt
```

# Loading The Data

```python
data=pd.read_csv('G:/LetsUpgrade/AI-ML/Day-27/train.csv')
```

```python
data.head(1)
```

```python
le=preprocessing.LabelEncoder()
```

```python
le.fit(data['Sex'])
```

```python
data['Sex']=le.transform(data['Sex'])
```

```python
data.isnull().sum()
```

```python
le.fit(data['Embarked'])
data["Embarked"]=le.transform(data["Embarked"])
```

```python
y=data['Pclass']
X=data.drop(["Pclass","PassengerId","Ticket","Name","Cabin"],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
knn=neighbors.KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train).score(X_test,y_test)
```

```python
y_pred=knn.predict(X_test)
```

```python
c=confusion_matrix(y_test,y_pred)
print('Confusion Matrix :- \n',c)
```

```python
print("Accuracy score based on k=1 ",round(accuracy_score(y_test,y_pred,normalize=True)*100,4))
```

```python
ac=c[0,0]+c[1,1]
anc=c[0,1]+c[1,0]
acc_score_with_conf_matrix=ac/(ac+anc)
print("Accuracy score using confusion matrix ",round(acc_score_with_conf_matrix*100,4))
```

```python
acc_score=[] 
def trainall(k):
    print("|---------------------------------------------------------------------------------------------|\n")
    print(" RUNNING LOOP FOR K=",k,"\n")
    knn=neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train).score(X_test,y_test)
    y_pred=knn.predict(X_test)
    c=confusion_matrix(y_test,y_pred)
    print("Accuracy score based on k=",k," using accuracy score module \n",round(accuracy_score(y_test,y_pred,normalize=True)
    *100,4),"\n")
    ac=c[0,0]+c[1,1]
    anc=c[0,1]+c[1,0] 
    acc_score_with_conf_matrix=round((ac/(ac+anc))*100,4)
    print("Accuracy score using confusion matrix for k=",k,"\n",round(acc_score_with_conf_matrix*100,4))
    acc_score.append(acc_score_with_conf_matrix)
    print("\n|---------------------------------------------------------------------------------------------|")
```

```python
for i in range(1,623):
    trainall(i)
```
