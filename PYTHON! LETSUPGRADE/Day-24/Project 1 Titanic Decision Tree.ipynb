# Import the packages

```python
import pandas as pd 
import numpy as np
from sklearn import preprocessing
from sklearn import tree
```

# Loading Train Dataset

```python
train_data=pd.read_csv('C:/Python/Day-24/Dataset/train.csv')
```

```python
train_data.head()
```

# Checking for Null values

```python
train_data.isna().sum()
```

# Now we are using Label Encoder

```python
label_encoder = preprocessing.LabelEncoder() 
encode_sex=label_encoder.fit_transform(train_data['Sex']) #fit_transform work as coverting text value into numrical value
```

# Now we are going to use Tree Model so our dependent variable is categerised so we are using Classifier


```python
tree_model=tree.DecisionTreeClassifier()
predicators=pd.DataFrame([encode_sex,train_data['Age'],train_data['Fare']]).T
tree_model.fit(X=predicators,y=train_data["Survived"])
```

```python
with open ('Titanic_surview.dot','w') as f:
    f=tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f)
```

### In above code we created one file as Dtree1.dot & passed with one variable named f & in variable we stored using tree fucntion we epxport graphviz & here we used tree model &  feature's as Sex,Age,Fare & used out_file for stroing the tree data in that location


# here we are going to calcualte Score of our model

```python
tree_model.score(X=predicators,y=train_data['Survived'])
```
